```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('qatar airways.csv')
display(df)
df.info()
```

```python
graph = sns.lineplot(x='Year', y='Passengers carried (in 000)', data=df, marker='o')
plt.xticks([2016, 2017, 2018])
plt.title('Qatar Airways')
plt.savefig('Qatar Airways.png')
```
